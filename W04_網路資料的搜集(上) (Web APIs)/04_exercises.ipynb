```python
from apiclient.discovery import build
import json

youtube = build('youtube', 'v3', developerKey=DEVELOPER_KEY)

result = youtube.videos().list(
    part='snippet,contentDetails,statistics',
    chart='mostPopular',
    regionCode='TW',maxResults=50).execute()

title = []
channelTitle = []
categoryId = []
viewCount = []
likeCount = []
dislikeCount = []
commentCount = []
    
for i in range(50):
        title.append(result['items'][i]['snippet']['title']) 
        channelTitle.append(result['items'][i]['snippet']['channelTitle'])
        categoryId.append(result['items'][i]['snippet']['categoryId'])
        viewCount.append(result['items'][i]['statistics']['viewCount'])
        likeCount.append(result['items'][i]['statistics']['likeCount'])
        dislikeCount.append(result['items'][i]['statistics']['dislikeCount'])
        commentCount.append(result['items'][i]['statistics']['commentCount'])

youtube_dict = {'title':title, 'channelTitle': channelTitle,'categoryId':categoryId,
                'viewCount':viewCount,'likeCount':likeCount,'dislikeCount':dislikeCount, 'commentCount':commentCount}
```

```python
import pandas as pd
df = pd.DataFrame(youtube_dict)
df
```

```python
import matplotlib.pyplot as plt
```

```python
category = pd.DataFrame(df['categoryId'].value_counts().reset_index())
category.columns =['Id','num']
```

```python
plt.figure(figsize=(6,9))
plt.pie(category["num"],
        labels = category["Id"],
        autopct = "%1.1f%%",
        pctdistance = 0.6,
        textprops = {"fontsize" : 12},
        shadow=False)
plt.axis('equal')
plt.title("Pie chart of category Id", {"fontsize" : 18})
plt.legend(loc = "best")
plt.show()
```

由以上可看出，政治新聞類(25)為發燒影片大宗，共有22支影片，佔據50名的幾乎一半，且普遍排名都靠前，這可能與剛結束的陳柏惟罷免案有著密不可分的關係。

其次為娛樂類(24)影片，共有11支影片。根據早前Ipsos針對台灣的Youtube使用族群做的調查指出，台灣觀眾的影片選擇取向主要以「放鬆」為主，且娛樂頻道的訂閱人數也僅次於音樂類頻道。

### 政治新聞類影片的特徵

有鑑於政治新聞類影片為大宗，不免讓人好奇是否因昨日的罷免案受到社會關注才有此效果，這裡看一下陳柏惟的事到底有多少媒體報導，其背後立場又為何。

```python
PNdf = df[df['categoryId'].isin(['25'])]
PNdf2 = PNdf['title'].str.contains("柏惟","刪Q")
print(PNdf2.value_counts())
```

在22則政治新聞當中，便有15則(幾乎是$\frac{2}{3}$)為陳柏惟罷免案相關新聞，可見此事足以作為新聞吸引閱聽人。
